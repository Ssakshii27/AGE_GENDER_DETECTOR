```python
img3=cv2.imread("wallpaperflare.com_wallpaper (8).jpg")
plt.imshow(img3)
```

```python
hsv=cv2.cvtColor(img3,cv2.COLOR_BGR2HSV)
figure_size=7
blur=cv2.blur(img3,(figure_size,figure_size))
plt.figure(figsize=(11,6))
plt.subplot(121),plt.imshow(cv2.cvtColor(hsv,cv2.COLOR_HSV2RGB)),plt.title("Original")
plt.subplot(122),plt.imshow(cv2.cvtColor(blur,cv2.COLOR_HSV2RGB)),plt.title("Blurred")
plt.xticks([]),plt.yticks([])
plt.show()
```

```python
img1=cv2.imread("7220885.jpg")
plt.imshow(img1)
```

```python
(height,width)=img1.shape[0:2]
zoom=cv2.resize(img1,None,fx=1.5,fy=1.5,interpolation=cv2.INTER_CUBIC)
plt.imshow(zoom)
```

```python
zoom=cv2.resize(img1,(int(width*2),int(height*2)),fx=1.5,fy=1.5,interpolation=cv2.INTER_CUBIC)
plt.imshow(zoom)
```

```python
m=cv2.getRotationMatrix2D((width/2,height/2),30,1)
rotate=cv2.warpAffine(img1,m,(height,width))
plt.imshow(rotate)
```
